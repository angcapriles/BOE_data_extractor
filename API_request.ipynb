```
import requests
from pandas.io.json import json_normalize
import pandas as pd
from google.colab import files

from_timestamp = '1546794000'

to_timestamp = '1546794000'

brand_id = 'CHEWYJUNIOR'



url = 'https://dw.ipos.vn/api/v1/modules/get-sale-manager?brand_id={brand}&from_timestamp={from_timestamp}&to_timestamp={to_timestamp}'.format(brand = brand_id,from_timestamp= from_timestamp, to_timestamp = to_timestamp)


session = requests.Session()
# these are sent along for all requests
session.headers['access_token'] = '3KGY640VB5ENO9VADMDXGZXAJM2DWQGB7XM8'
session.headers['secret_key'] = '658b88e33f2cdd3b03880b11c3823f54'
# not strictly needed, but the documentation recommends it.
session.headers['Accept'] = "application/json; charset=UTF-8"

# log in first, to get the tokens
response = session.get(
    url
)

data = response.json()

df1 = json_normalize(data, 'data')

d = []

for row_index,row in df1.iterrows():
    for r in row['sale_detail']:
        dn = {}
        for key in df1.keys():
            if key != 'sale_detail' and key != 'sale_payment_method':
                dn[key] = row[key]
        dn['item_id'] = r['item_id']
        dn['price_sale'] = r['price_sale']
        dn['quantity'] = r['quantity']
        
        for pay in row['sale_payment_method']:
            if pay['list_order'] == r['list_order']:
                dn['payment_type'] = pay['payment_method_id']
            else:
                dn['payment_type'] = 'N/A'
        d.append(dn)
print(d)  
df = pd.DataFrame.from_records(d)
filename = url.split('brand_id=')[1].split('&')[0] + '.csv'
df.to_csv(filename , sep=',', encoding='utf-8', index=False)
files.download(filename)
```
